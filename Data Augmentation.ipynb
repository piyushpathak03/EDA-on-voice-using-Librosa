## Convert 1 wav to multiple


```python
from pydub import AudioSegment
from pydub.utils import make_chunks

myaudio = AudioSegment.from_file("normal breath sound (8).wav" , "wav") 
chunk_length_ms = 5000 # pydub calculates in millisec
chunks = make_chunks(myaudio, chunk_length_ms) #Make chunks of one sec

#Export all of the individual chunks as wav files

for i, chunk in enumerate(chunks):
    chunk_name = "chunk{0}.wav".format(i)
    print("exporting", chunk_name)
    chunk.export(chunk_name, format="wav")
```

```python
import mutagen
from mutagen.wave import WAVE
import wave
import contextlib
import csv
import os
from glob import glob
from pydub import AudioSegment 
from pydub.utils import make_chunks
import os



DIRNAME = r'C:\Users\hp\Desktop\Voice disease prediction\data\Raw_data\Wheezing'
OUTPUTFILE = r'C:\Users\hp\Desktop\Voice disease prediction\data\Chuncked_data\chuncked_Wheezing'

def process_sudio(file_name):
    myaudio = AudioSegment.from_file(file_name, "wav") 
    chunk_length_ms = 7500 # pydub calculates in millisec 
    chunks = make_chunks(myaudio,chunk_length_ms) #Make chunks of one sec 
    return chunks


for root, dir, files in os.walk(DIRNAME):
    for file in files:
        (filepath, ext) = os.path.splitext(file)
        os.chdir(DIRNAME)
        file_name = os.path.basename(file)          
#         print(filepath)
        if ext == '.wav':

### checking the duration of audio
            with contextlib.closing(wave.open(file,'r')) as f:
                frames = f.getnframes()
                rate = f.getframerate()
                duration = frames / float(rate)
#                 print(duration)
                if duration>16:
                    print("Duration of {0} is {1}".format(filepath, duration))
## processing the audio into chunks
                    a = process_sudio(file)
                else:
                    print('no file')
## Saving back the chuncks                    
                for i, chunk in enumerate(a):
                    chunk_name = filepath+ '_Chuncked' + "_{0}.wav".format(i)
                    print ("exporting", chunk_name)
                    chunk.export(chunk_name, format="wav")            
```
